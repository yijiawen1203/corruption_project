**Required output_attr_stats.csv, top_organization.json
**Generates top_org_email.csv**  

```python
import csv
import pandas as pd
from tqdm import tqdm_notebook, tnrange
from datetime import datetime
import numpy as np
import json
```

# load email and top_organization data

```python
df = pd.read_csv('output_attr_stats.csv')
from_list = df['From'].values.tolist()
to_list = df['To'].values.tolist()
cc_list = df['CC'].values.tolist()
length = len(df)
```

```python
with open('top_organization.json', 'r') as f:
    top_organization_dict = json.load(f)
```

```python
top_org_list = []
for v in top_organization_dict.values():
    top_org_list += v
```

```python
len(top_organization_dict.keys())
```

# Select all emails between top organization people

```python
#get the index of the related email
idx_list = []
for i in range(length):
    if not pd.isnull(from_list[i]) and from_list[i] in top_org_list:
        if not pd.isnull(to_list[i]):
            recipients = to_list[i].split(';')
            for j in range(len(recipients)):
                if recipients[j] in top_org_list:
                    idx_list.append(i)
                    #break
        if not pd.isnull(cc_list[i]):
            recipients = cc_list[i].split(';')
            for j in range(len(recipients)):
                if recipients[j] in top_org_list:
                    idx_list.append(i)
                    #break
```

```python
idx_list = list(set(idx_list))
```

```python
df_target = df.iloc[idx_list]
df_target = df_target.reset_index(drop=True)
df_target
```

```python
df_target.to_csv('top_organization_email.csv',index=0)
```

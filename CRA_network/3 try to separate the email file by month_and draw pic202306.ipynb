# 根据已经有的adj matrix，计算individual reciprocity

```python
#test for one individual reciprocity

test_path='/Users/yoga/Documents/Projects/4 合作项目 腐败 economic corruption/68_update for 20220429/adj_matrix by month/22adjmatrix_2016-05.csv'
data=pd.read_csv(test_path,index_col=0)
name=data.columns.to_list()
adj_s1 = data.to_numpy() #adjacent matrix

#creat weighted edges from matrix
weighted_edge_list=[]
for i in range(len(name)):
#     print(i)
    for j in range(len(name)):
        if adj_s1[i,j]!=0:
            edgeij=(name[i],name[j],adj_s1[i,j])
            weighted_edge_list.append(edgeij)
            
#create graph

# Create a directed graph (DiGraph) using NetworkX
G = nx.DiGraph()

# Add weighted edges to the graph
G.add_weighted_edges_from(weighted_edge_list)

# Calculate reciprocity for each individual node
node2reciprocity={}
for node in G.nodes():
#     print(node)
    out_edges = set(G.successors(node))
    in_edges = set(G.predecessors(node))
    reciprocal_edges = out_edges.intersection(in_edges)
    total_out_edges = len(out_edges)
    if total_out_edges > 0:
        reciprocity = len(reciprocal_edges) / total_out_edges
    else:
        reciprocity = 0
    print(f"Reciprocity for  {node}: {reciprocity}")
    
    node2reciprocity[node]=reciprocity

#put dictionary into a dataframe
pd_node2reciprocity = pd.DataFrame(list(node2reciprocity.items()), columns=['people involving in email', 'reciprocity'])

pd_node2reciprocity['time']=''

time = '2016-05'
pd_node2reciprocity['time']=time





```

```python
#calculate for all node reciprocity for all month
#尝试计算reciprocity
#已经确认 adj matrix 是不对称的，是有方向的
adj_folder = r'/Users/yoga/Documents/Projects/4 合作项目 腐败 economic corruption/68_update for 20220429/adj_matrix by month'
all_files = glob.glob(adj_folder + "/*.csv")
outdir = '/Users/yoga/Documents/Projects/4 合作项目 腐败 economic corruption/65_update_the graph_of_joey_and_yijia_in_paper_202206/1_yijia_small netework data/individual_recipricity_20230606'

try:
    os.makedirs(outdir)
except FileExistsError:
    # directory already exists
    pass

ind=0

whole_network=[]

for file in all_files:
#     if ind>0:
#         break
    tmp = file.split('/')
    time = tmp[-1].replace('.csv','').split('_')[-1]
    ind=ind+1
    
    print(ind,':',time)
   
    data=pd.read_csv(file,index_col=0)
    name=data.columns.to_list()
    adj_s1 = data.to_numpy() #adjacent matrix

    #creat weighted edges from matrix
    weighted_edge_list=[]
    for i in range(len(name)):
    #     print(i)
        for j in range(len(name)):
            if (i!=j)& (adj_s1[i,j]!=0):
                edgeij=(name[i],name[j],adj_s1[i,j])
                weighted_edge_list.append(edgeij)

    #create graph

    # Create a directed graph (DiGraph) using NetworkX
    G = nx.DiGraph()

    # Add weighted edges to the graph
    G.add_weighted_edges_from(weighted_edge_list)

    # Calculate reciprocity for each individual node
    node2reciprocity={}
    for node in G.nodes():
    #     print(node)
        out_edges = set(G.successors(node))
        in_edges = set(G.predecessors(node))
        reciprocal_edges = out_edges.intersection(in_edges)
        total_out_edges = len(out_edges)
        if total_out_edges > 0:
            reciprocity = len(reciprocal_edges) / total_out_edges
        else:
            reciprocity = 0
        print(f"Reciprocity for  {node}: {reciprocity}")

        node2reciprocity[node]=reciprocity

    #put dictionary into a dataframe
    pd_node2reciprocity = pd.DataFrame(list(node2reciprocity.items()), columns=['people involving in email', 'reciprocity'])

    pd_node2reciprocity['time']=''

    pd_node2reciprocity['time']=time
    
    whole_network.append(pd_node2reciprocity)




    
    
```
